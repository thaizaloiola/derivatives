```python
#Estratégia de Straddle
```

```python
#Stock price

preco_spot = 10.65

#compra put straddle

strike_compra_put = 10.48

#preço put straddle

premio_compra_put =0.34

#compra callstraddle

strike_compra_call = 10.48

#preço call straddle 

premio_compra_call= 0.50

#range de preços que a ação pode assumir na data de expiração

sT = np.arange (0, 2*preco_spot,1)

sTT =np.arange (0, 2*preco_spot,1)
```

```python
#Payoff da call

def call_payoff(sT, strike_preco, premio):
    return np.where(sT > strike_preco, sT - strike_preco, 0) - premio
payoff_compra_call = call_payoff (sT, strike_compra_call, premio_compra_call)
```

```python
#montando o gráfico do payoff da call

fig, ax = plt.subplots ()

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_position('zero')
ax.plot(sT, payoff_compra_call, label='Compra call', color='r')
plt.xlabel('Preço ITSA4')
plt.ylabel('Payoff')
plt.legend()
plt.show()
```

```python
#Payoff da put

def put_payoff (sTT, strike_preco, premio):
    return np.where(sTT < strike_preco, strike_preco - sTT, 0)-premio

payoff_compra_put = put_payoff (sTT, strike_compra_put, premio_compra_put)
```

```python

#montando o gráfico do payoff da put

fig, ax = plt.subplots()

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_position('zero')
ax.plot(sTT, payoff_compra_put, label='Compra put', color ='g')
plt.xlabel('Preço ITSA4')
plt.ylabel('Payoff')
plt.legend()
plt.show()
```

```python
payoff_straddle = payoff_compra_call + payoff_compra_put
```

```python
print ("Lucro máximo: ilimitado")
print ("Perda máxima:", min(payoff_straddle))

fig, ax = plt.subplots()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_position('zero')

ax.plot(sT, payoff_compra_call, '--', label= 'Compra call', color ='r')
ax.plot(sTT, payoff_compra_put, '--', label= 'Compra put', color= 'g')

ax.plot(sT, payoff_straddle, label='Straddle')
plt.xlabel('Preço ITSA4', ha='left')
plt.ylabel('Profit and loss')
plt.legend()
plt.show()


```

```python
#Stock price

preco_spot_stg = 10.65

#compra put strangle

strike_compra_put_stg = 10.48

#preço put strangle

premio_compra_put_stg = 0.34

#compra call strangle

strike_compra_call_stg = 11.73

#preço call strangle 

premio_compra_call_stg= 0.07

#range de preços que a ação pode assumir na data de expiração

sTTT =np.arange (0.8*preco_spot_stg, 1.3*preco_spot_stg,0.01)
```

```python
#Payoff da put

def put_payoff (sTTT, strike_preco, premio):
    return np.where(sTTT < strike_preco, strike_preco - sTTT, 0) - premio

payoff_compra_put_stg = put_payoff (sTTT, strike_compra_put_stg, premio_compra_put_stg)
```

```python
# Plot
fig, ax = plt.subplots()
ax.spines['bottom'].set_position('zero')
ax.plot(sTTT,payoff_compra_put_stg,label='Compra Put',color='g')
plt.xlabel('Preço ITSA4')
plt.ylabel('Payoff')
plt.legend()
plt.show()
```

```python
#Payoff da call

def call_payoff (sTTT, strike_preco, premio):
    return np.where(sTTT > strike_preco, - strike_preco + sTTT, 0) - premio

payoff_compra_call_stg = call_payoff (sTTT, strike_compra_call_stg, premio_compra_call_stg)
```

```python
# Plot
fig, ax = plt.subplots()
ax.spines['bottom'].set_position('zero')
ax.plot(sTTT,payoff_compra_call_stg,label='Compra Put',color='g')
plt.xlabel('Preço ITSA4')
plt.ylabel('Payoff')
plt.legend()
plt.show()
```

```python
payoff_strangle = payoff_compra_call_stg + payoff_compra_put_stg
```

```python
#Gráfico Strangle

print ("Lucro máximo: ilimitado")
print ("Perda máxima:", min(payoff_strangle))

fig, ax = plt.subplots()

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_position('zero')

ax.plot(sTTT, payoff_compra_call_stg, '--', label= 'Compra call', color ='r')
ax.plot(sTTT, payoff_compra_put_stg, '--', label= 'Compra put', color= 'g')

ax.plot(sTTT, payoff_strangle, label='Strangle')
plt.xlabel('Preço ITSA4', ha='left')
plt.ylabel('Profit and loss')
plt.legend()
plt.show()
```

```python
#Gráfico vende call

fig, ax = plt.subplots()
ax.spines['bottom'].set_position('zero')
ax.plot(sT_bcs,payoff_vende_call_bcs,label='Vende call Strike 11,73',color='r')
plt.xlabel('Preço ITSA4')
plt.ylabel('Profit and loss')
plt.legend()
plt.show()
```

```python
#preço inicial da ação

s0_ppp= 10.65

sT_ppp = np.arange(0,s0_ppp*2,5)

k=10.48
```

```python
y1 = np.where( sT_ppp > k, (sT_ppp - k),0)
y2 = np.where(sT_ppp <k, (k - sT_ppp),0)
cash = np.zeros(len(sT_ppp))+k
```

```python
def graph(text, text2=''): 
    pl.xticks(())
    pl.yticks(())
    pl.xlim(0,20)
    pl.ylim(0,20)
    pl.plot([k,k],[0,3])
    pl.text(k,0,"X")
    pl.text(0,k,"X")
    pl.text(k,k*1.7, text, ha= 'center', va='center', size =10, alpha=.5)
    pl.text(-3.5,10,text2,size=16)
    
pl.figure(figsize=(6,4))
pl.subplot(2,3,1); graph('Payoff da call'); pl.plot(sT_ppp,y1)
pl.subplot(2,3,2); graph('Cash','+');       pl.plot(sT_ppp, cash)
pl.subplot(2,3,3); graph('Porfolio A','='); pl.plot(sT_ppp, cash+y1)
pl.subplot(2,3,4); graph('Payoff da Put');  pl.plot(sT_ppp, y2)
pl.subplot(2,3,5); graph('Ação da ITSA4','+');      pl.plot(sT_ppp, sT_ppp)
pl.subplot(2,3,6); graph('Porfolio B','='); pl.plot(sT_ppp, sT_ppp+y2)
pl.show()
    
```

```python

#preços dos prêmios da call e da put respectivamente:

premio_call =  0.5

premio_put = 0.34

# preço ativo objeto hoje

s0_ppp = 10.65

#taxa livre de risco = valor % aproximado da t-bond de 30 anos do tesouro americano

rf = 0.02

#tempo restante até o vencimento da opção em dias úteis como percentual do ano 

t= 27/252

#operação em taxa livre de risco

oper_rf = 10.48 / ((1+ rf)**t)

# S0 + prêmio da put de mesmo strike e prazo que a call:

s0_premioput = premio_put + s0_ppp

#premio da call de mesmo strike e prazo que a put + o valor da operação em taxa livre de risco 

premiocall_oper = premio_call + oper_rf

```

```python
import pandas as pd
import numpy as np
import scipy as sp

from numpy import sqrt,mean,log,diff

from scipy import stats
```

```python
#Código para o cálculo da Volatilidade implícita

def bsm_price(option_type, sigma, s, k, r, T, q):
    # calcular o preço via Black-Scholes das opções europeias de compra e venda
    sigma = float(sigma)
    d1 = (np.log(s / k) + (r - q + sigma ** 2 * 0.5) * T) / (sigma * np.sqrt(T))
    d2 = d1 - sigma * np.sqrt(T)
    
    if option_type == 'c':
        
        price = np.exp(-r*T) * (s * np.exp((r - q)*T) * stats.norm.cdf(d1) - k *  stats.norm.cdf(d2))
        return price
    
    elif option_type == 'p':
        
        price = np.exp(-r*T) * (k * stats.norm.cdf(-d2) - s * np.exp((r - q)*T) *  stats.norm.cdf(-d1))
        return price
    
    else:
        
        print('No such option type %s') %option_type
        
def implied_vol(option_type, option_price, s, k, r, T, q):
    # aplicar o método de bissecção para obter a volatilidade implícita, resolvendo a função BSM
    precision = 0.00001
    upper_vol = 500.0
    max_vol = 500.0
    min_vol = 0.0001
    lower_vol = 0.0001
    iteration = 0

    while 1:
        
        iteration +=1
        mid_vol = (upper_vol + lower_vol)/2.0
        price = bsm_price(option_type, mid_vol, s, k, r, T, q)
        
        if option_type == 'c':

            lower_price = bsm_price(option_type, lower_vol, s, k, r, T, q)
            
            if (lower_price - option_price) * (price - option_price) > 0:
                lower_vol = mid_vol
                
            else:
                upper_vol = mid_vol
                
            if abs(price - option_price) < precision: break 
                
            if(mid_vol > max_vol - 5) :
        
                mid_vol = 0.000001
                
                break

        elif option_type == 'p':
            upper_price = bsm_price(option_type, upper_vol, s, k, r, T, q)

            if (upper_price - option_price) * (price - option_price) > 0:
                upper_vol = mid_vol
            else:
                lower_vol = mid_vol
                
            if abs(price - option_price) < precision: break 
                
            if iteration > 50: break
                
    # a volatilidade implícita da opção de compra europeia (com prêmio c = 0,5, S0 = 10,65, k = strike = 10,48, r = 0,02, T = 27/252, d = 0,01)

    return mid_vol
implied_vol('c', 0.5, 10.65, 10.48, 0.02, 27/252, 0.01)
```
